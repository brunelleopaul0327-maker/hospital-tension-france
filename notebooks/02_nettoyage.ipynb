```python
import pandas as pd

# Charger le fichier brut
df = pd.read_csv("data/raw/covid-hosp-txad-reg-2023-06-30-16h29.csv", sep=";")

print(f"Taille brute : {df.shape}")
print(df.dtypes)
```

```python
# Filtrer PourAvec = 0
df = df[df["PourAvec"] == 0]
print(f"Taille après filtre : {df.shape}")
```

```python
# Convertir la colonne jour en datetime
df["jour"] = pd.to_datetime(df["jour"])

print(df["jour"].dtype)
print(df["jour"].min(), "→", df["jour"].max())
```

```python
# Valeurs manquantes par colonne
print(df.isnull().sum())
print(f"\nSoit en pourcentage :")
print((df.isnull().sum() / len(df) * 100).round(2))
```

```python
# Trier par région puis par date
df = df.sort_values(["reg", "jour"]).reset_index(drop=True)

# Extraire des informations temporelles
df["annee"] = df["jour"].dt.year
df["mois"] = df["jour"].dt.month
df["semaine"] = df["jour"].dt.isocalendar().week.astype(int)

print(df.head())
print(df.shape)
```

```python
# Sauvegarder en Parquet
df.to_parquet("data/processed/hopitaux_clean.parquet", index=False)
print("✓ Fichier sauvegardé dans data/processed/hopitaux_clean.parquet")
```
